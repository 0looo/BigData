#####  2010~2019 년도 출입파일을 앞선 전처리한 후 하나의 df로 통합

```python
import os
```

```python
df_list = []
for i in os.listdir('./tourist_files')[:-1]:
    excel = './tourist_files'+i
    df = pd.read_excel('./tourist_files/kto_201901.xlsx',header = 1,usecols ='A:G', skipfooter = 4)
    df['기준년월'] = i.split('.xlsx')[0].split('_')[1]
    continent_list = ['아시아주','아시아 기타','국적미상','미주','구주','대양주','아프리카주','아프리카 기타','기타대륙','교포소계','대양주 기타','구주 기타','미주 기타']
    df=df[df.국적.isin(continent_list) == False]      # bool == False 하면 반대로 변경됨.
    df.reset_index(drop = True, inplace = True)
    
    df_list.append(df)
df_list
```

```python
pd.concat(df_list)
```

```python
df = df_list[0]
for i in df_list[1:] :
    df = pd.concat([df,i])
df
```

##### df2015 ~ df2019 모든 파일을 동일한 방식으로 변형해서 하나의 dataframe으로 합하라.

```python
l = []
for i in range(2015,2020) : 
    csv = './data/ad/'+str(i)+'ad.csv'
    df = pd.read_csv(csv)
    df = df.drop('Unnamed: 0',axis=1)
    df = df.drop(0)
    df.set_index('firm',inplace = True)
    df['tv'] = df['tv'].astype('int')
    df['radio'] = df['radio'].astype('int')
    df['newpaper'] = df['newpaper'].astype('int')
    df['magazine'] = df['magazine'].astype('int')        # index설정, 각 항목 int설정 완료(년도별)
    df['total'] = df.sum(axis=1)
    df['date'] = str(i) + '09'              # total, date의 col추가
    l.append(df)             # list에 dataframe 넣음
l
```

```python
df = pd.concat(l)      # list를 concat하면 편리함.
df.head()
```

###  col별 날짜형식으로 변경법
    pd.to_data_datetime(df['date']. format=%Y%m)   # 원본에 영향주지 않음

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m')
```

```python
df['year']= df.date.dt.year

df.head()
```

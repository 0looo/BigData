##### 여러페이지 크롤링

```python
driver = webdriver.Chrome('D:\\BigData\\crawling\\chromedriver.exe')
total = []
page = 10

for j in tqdm_notebook(range(1,page+1)):
    url =  'http://search.danawa.com/dsearch.php?query=%26%2339%3B%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0%26%2339%3B&originalQuery=%26%2339%3B%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0%26%2339%3B&volumeType=allvs&page={}&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&tab=goods'.format(j)
    driver.get(url)
    time.sleep(2)
    
    html = driver.page_source
    soup = bs(html,'html.parser')
    prod_item = soup.select('div.main_prodlist > ul > li')
    for i in prod_item:
        total.append([i.select('p.prod_name>a')[0].text.strip(),i.select('div.spec_list')[0].text.strip(),i.select('p.price_sect > a')[0].text.strip()])
total
```

##### 원하는 스팩 추출, 필요없는 col삭제 및 col type 조정

```python
import re
```

```python
spec = []  #스팩
h = []     #흡입력

for i in df.스팩:
    try :
        spec.append(re.compile('[사][용][시][간]\:\s\d+[ㄱ-힣]+').findall(i)[0])
    except :
        spec.append(np.nan)
    try:
        h.append(re.compile('[흡][입][력]+\:\s\d+\,?\d+[a-zA-Z]+').findall(i)[0])
    except :
        h.append(np.nan)
```

```python
del df['스팩']
```

```python
df['사용시간'] = spec
df['흡입력'] = h

df.head()
```

```python
df['가격'] = df['가격'].apply(lambda x: int(x.split('원')[0].replace(',','')))
df.head()
```

```python
hour = []
for i in df.사용시간:
    try :
        i = i.split('사용시간:')[-1]

        if '분' in i :
            hour.append(int(re.compile('[0-9]+').findall(i)[0]))
        elif '시간' in i :
            hour.append(int(re.compile('[0-9]+').findall(i)[0])*60)
    except:
        hour.append(np.nan)
hour
```

```python
df['사용시간'] = hour
```

```python
def get_sucktion(value):
    try:
        value = value.split(': ')[-1]
        value = value.upper()
        if 'AW' in value or 'W' in value:
            result = value.replace('A','').replace('W', '')
            result = int(result.replace(',',''))
        elif 'PA' in value:
            result = value.replace('PA','')
            result = int(result.replace(',',''))/100
        else:
            result = None
        return result
    except:
        return np.nan
    
```

```python
df['흡입력'] = df['흡입력'].apply(get_sucktion)
```

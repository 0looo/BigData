###  BeautifulSoup parsing
#### parsing
* html tag 안에 있는 내용을 추출하는 것
* parsing 모듈 : BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
html = """
    <head>
        <title> 파이썬 크롤링 </title>
    </head>
  
    <body>
    <table border = 1>
        <tr>
            <td> 항목 </td> 
            <td> 2013 </td>
            <td> 2014 </td>
        </tr>
        <tr>
            <td> 매출액 </td>
            <td> 100 </td>
            <td> 200 </td>
        </tr>
    </table>
    </body>"""
```

```python
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.select('td')
```

```python
soup.select('tr')
```

```python
soup.select('td')[1].text   # .text 는 값만 가져옴
```

```python
for i in soup.select('td') :
    print(i.text)
```

###  naver main에서 main menu 추출

```python
from bs4 import BeautifulSoup
import requests  # 웹사이트의 html script를 긁어오는 라이브러리

url = 'https://www.naver.com/'

html = requests.get(url)
print(html)
print(html.text)
```

```python
#NM_FAVORITE > div.group_nav > ul.list_nav.NM_FAVORITE_LIST > li
```

```python
soup = BeautifulSoup(html.text,'html.parser')
```

```python
res = soup.select ("#NM_FAVORITE > div.group_nav > ul.list_nav.NM_FAVORITE_LIST > li > a")
res
```

```python
for t in res :
    print(t.text)
```

```python
for t in res : 
    print(t['href'])
```

```python
for t in res :
    print(t.text,';',t['href'])
```

```python
#newsColl > div.coll_cont
```

```python
for t in res : 
    print(t['href'])
```

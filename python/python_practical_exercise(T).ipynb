```python
def thousand_separator(): 
    num = input('길이가 5 이상인 임의의 양의 정수를 입력해주세요.')
    num[::-1]

    new_num = ''

    for i, n in enumerate(num[::-1]):
        if (i%3 == 0) & (i != len(num)) & (i != 0):
            new_num += ','+n
        else:
            new_num += n

    return new_num[::-1]       
```

```python
with open('poem.txt') as f:
    data = f.read()
    d = {}
    for c in data:
        if c == ' ':
            continue  
        elif c in d:
            d[c] += 1
        else:
            d[c] = 1
sorted(d.items(), key = lambda x : x[1], reverse=True)      
```

```python
# 텍스트 파일을 열어서 특정 단어가 몇 개 존재하는지 출력.
# 단, count를 사용하지 않는다..
```

```python
def word_count(file, word):
    count = 0
    with open(file, encoding='utf-8') as f:
        data = f.read()
        idx = data.find(word)
        while idx != -1:
            count += 1
            idx = data.find(word, idx+1)
            print(idx)
    return count
```

```python
# # 연도별 남아 신생아 수와 여아 신생아 수를 각각 출력..
```

```python
path = "D:\\BigData\\python\\names\\"
file_path = path + "yob{}.txt"

for y in range(1880, 2016):
    file_path1 = file_path.format(y)
    
    with open(file_path1, 'r') as f:
        data = f.readlines()
        fbs = 0
        mbs = 0
        for l in data:
            bp = l.split(',')[2]
            bg = l.split(',')[1]
            if bg == 'F':
                fbs += int(bp)
            else:
                mbs += int(bp)
    print("년도 : {}, 남아출생아수: {}, 여아출생아수: {}".\
          format(y, mbs, fbs))
```

```python
## 연도별 인기있는 상위 10개 이름을 성별로 출력..
```

```python
path = "D:\\BigData\\python\\names\\"
file_path = path + "yob{}.txt"

for y in range(1880, 2016):
    file_path1 = file_path.format(y)
    
    with open(file_path1, 'r') as f:
        data = f.readlines()
        fbn = {}
        mbn = {}
        for l in data:
            bp = l.split(',')[2]
            bg = l.split(',')[1]
            bn = l.split(',')[0]

            if bg == 'F':
                fbn[bn] = int(bp)
            else:
                mbn[bn]= int(bp)
                
        fbs = sorted(fbn.items(), key=lambda x: x[1], reverse=True)
        mbs = sorted(mbn.items(), key=lambda x: x[1], reverse=True)
        
        fns = []
        for i in fbs[:10]:
            fns.append(i[0])
            
        mns = []
        for i in mbs[:10]:
            mns.append(i[0])
            
    print('연도:', y)
    print('남아 10:', mns)
    print('여아 10:', fns)
    print('='*30)
 
```

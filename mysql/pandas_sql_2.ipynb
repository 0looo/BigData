### 'sakila' 에  'rental'테이블에서 월별 렌탈 수를 출력
##### 날짜는 오름차순으로 출력

```python
# pandas
df = pd.read_sql('select * from rental', db)
df.info()
```

```python
df['month'] = df.rental_date.dt.month
df.groupby('month')['rental_id'].count()
```

```python
# sql
pd.read_sql('select month(rental_date), count(rental_id) "횟수" from rental group by month(rental_date)', db)
```

##### 'payment' table엣 월별 전체매출, 평균매출, 결제건수

```python
# pandas
df = pd.read_sql('select * from payment',db)
df
```

```python
df['month'] = df.payment_date.dt.month
df.groupby('month').agg({'amount':[np.mean,sum],'payment_id':'count'})
```

```python
# sql
pd.read_sql("""
select month(payment_date) month, avg(amount) 평균매출, sum(amount) 총매출, count(payment_id) 매출건수 
from payment group by month;
""",db)
```

#### 속도비교

```python
import time
```

##### ex1 : name, population, continent에서 contient = europe, population 10,000,000이상

```python
# sql ex1
start_time = time.time()

pd.read_sql("""
select name, population, continent from country 
where continent = 'europe' and population >= 10000000;
""",db)

end_time = time.time()

execute_time = end_time- start_time
execute_time
```

```python
# pandas ex1
start_time = time.time()

df = pd.read_sql("select * from country;",db)[['Name','Population','Continent']]
df[(df.Continent == 'Europe') & (df.Population >= 10000000)]

end_time = time.time()

execute_time = end_time- start_time
execute_time
```

#####  ex2) continent가 europe or asia

```python
# sql ex2
start_time = time.time()

pd.read_sql("""
select * from country 
where continent in ('europe', 'asia');
""",db)

end_time = time.time()

execute_time = end_time- start_time
execute_time
```

```python
# pandas ex2
# sql ex2
start_time = time.time()

df = pd.read_sql("""
select * from country
""",db)
df[df.Continent.isin(['Europe','Asia'])]

end_time = time.time()

execute_time = end_time- start_time
execute_time
```

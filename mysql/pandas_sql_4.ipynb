```python
# sql
pd.read_sql('''select c.customer_id, sum(amount) user별매출 from customer c join payment t 
where c.customer_id = t.customer_id 
group by c.customer_id
;''', db)
```

```python
#pandas
cus = pd.read_sql('select * from customer', db)
cus
```

```python
pay = pd.read_sql('select * from payment', db)
pay
```

```python
pd.merge(cus,pay, on = 'customer_id').groupby('customer_id')[['amount']].sum()
```

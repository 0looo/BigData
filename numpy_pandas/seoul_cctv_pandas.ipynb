```python
pop_seoul = pd.read_excel('./data/population_in_Seoul.xls',usecols = [1,3,6,9,13],skiprows=2)
pop_seoul.head()
```

```python
# columns 이름 변경 -> 원본은 변경되지 않음
# 원본변경을 원할시 ,inplace=True를 안에 추가하면 됨
pop_seoul.rename(columns = {'자치구' : '구별','계': '인구수','계.1':'한국인','계.2':'외국인','65세이상고령자':'고령자'},inplace=True)
pop_seoul.head()
```

```python
pop_seoul.drop(26,axis=0, inplace= True)
pop_seoul.tail()
```

#### 구별 외국인 비율 상위 5개 구

```python
# 외국인 비율
pop_seoul['외국인비율'] = pop_seoul['외국인']/pop_seoul['인구수']*100
pop_seoul.sort_values('외국인비율',ascending = False).head().구별
```

####  cctv데이터 분석

```python
cctv_seoul = pd.read_csv('./data/CCTV_in_Seoul.csv')
cctv_seoul.head()
```

```python
cctv_seoul.info()
```

```python
cctv_seoul.rename(columns = {'기관명':'구별'}, inplace=True)
cctv_seoul.head()
```

####  cctv가 가장많은 구 상위 5곳

```python
cctv_seoul.sort_values('소계',ascending = False)[:5]
```

##### 2016년도에 cctv증가율이 가장 높은구?

```python
# 증가율
cctv_seoul['2016 증가율'] = cctv_seoul['2016년']/(cctv_seoul['소계']-cctv_seoul['2016년'])
cctv_seoul.sort_values('2016 증가율', ascending = False)[:5]
```

#### 1인당 cctv수(인당 cctv비율)이 높은 상위 5개 구 출력

```python
total_seoul = pd.merge(cctv_seoul,pop_seoul, on='구별',how = 'outer')
```

```python
total_seoul.head()
```

```python
total_seoul['인당 cctv수'] = total_seoul.소계 / total_seoul.인구수
total_seoul.sort_values('인당 cctv수', ascending = False)[:5]
```

#### 1인당 cctv에 가장 큰 영향을 미치는 요인은 ?

```python
total_seoul.corr()['인당 cctv수'].sort_values(ascending=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

```python
drinks = pd.read_csv(url)
drinks.head()
```

```python
# drinks.rename(columns = {'col': 'coll'})
drinks.columns = ['country','beer','spirit','wine','liters','continent']   # 전체 col 바꿀때 
```

```python
drinks.info()
```

```python
drinks.set_index('country',inplace = True)  # 'country' col을 index로 설정, / inplace -> 원본에 변경
```

## Line Chart

```python
plt.figure(figsize = (5,3))
drinks['beer'].plot()
```

```python
drinks.plot()
```

```python
plt.figure(figsize=(10,3))
drinks.sort_values('beer')['beer'].plot()
```

## bar chart
    y값에 크기를 표시

```python
plt.figure(figsize = (20,3))
drinks['beer'].plot(kind = 'bar')
```

```python
plt.figure(figsize = (20,3))
drinks['beer'].sort_values(ascending = False)[:100].plot(kind = 'bar')
```

#### 대륙별 평균 맥주소비량을 bar chart로 표시

```python
plt.figure(figsize = (4,3))
drinks.groupby('continent')['beer'].mean().plot(kind='bar')
```

#### 대륙별 모든 음료 평균을 bar chart로 표시

```python
plt.figure(figsize = (4,3))
drinks.groupby('continent').mean().plot(kind='bar')
```

```python
plt.figure(figsize = (5,3))
drinks.groupby('continent').mean().plot(kind='bar',stacked=True)
```

## histogram
    구간에 포함된 데이터의 수를 bar

```python
drinks['beer'].plot(kind='hist', bins = 50)
plt.xlabel('Beer Servings')
plt.ylabel('Frequency')
plt.title('Histogram of Beer')
```

## scatter
    x좌표와 y좌표 필요

```python
plt.figure(figsize = (5,3))
drinks.plot(kind='scatter', x = 'beer', y = 'wine')
```

```python
pd.plotting.scatter_matrix(drinks)  # 모든 col 간에 scatter 표시
```

## boxplot

```python
drinks.plot(kind = 'box')
```

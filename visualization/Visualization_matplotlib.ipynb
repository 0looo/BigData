### line plot

```python
plt.figure(figsize=(3,2))
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
```

```python
plt.figure(figsize = (3,2))

x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.figure(figsize = (3,2))

x1 = range(50)
y1 = [np.cos(value) for value in x1]

y2 = [np.sin(value) for value in x1]

plt.plot(x1,y1)
plt.plot(x1,y2)
```

```python
plt.figure(figsize=(4,3))

plt.subplot(1,2,1)      # subplot  :  도화지 나눔 (행, 렬, 번호)
plt.plot(x1, y1, color = 'red', linestyle = 'dotted')   # color : 그래프의 색지정
                                    # linestyle  : ex) dotted, dashed....

plt.subplot(1,2,2)
plt.plot(x1, y2, color = 'green',linestyle = 'dashed')
```

```python
plt.figure(figsize = (4,3))

x = np.random.rand(500)
y = x + np.random.randn(500)/5          # randn : 평균 1, var 1에서 랜덤추출

plt.scatter(x,y)
```

```python
data1 = np.random.rand(512,2)
data1
```

```python
data2 = np.random.rand(512,2)
data2
```

```python
plt.figure(figsize = (4,3))

x1 = data1[:,0]
y1 = data1[:,1]


x2 = data2[:,0]
y2 = data2[:,1]


plt.scatter(x1,y1, c='b', marker='x')
plt.scatter(x2,y2, c='g', marker = '^')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.random.rand(50)
area = np.pi*(30*np.random.rand(50))**2

plt.scatter(x,y, c=colors, s = area, alpha = 0.3)
```

### Bar Plot

```python
data = np.array([[5,25,52,20],[4,23,51,17],[6,22,52,19]])
data
```

```python
x = np.arange(0,8,2)
x
```

```python
y = data[1]
```

```python
plt.figure(figsize = (4,3))

plt.bar(x,y, color= 'g', width=0.5)
```

```python
data
```

```python
y0 = data[0]
y1 = data[1]
y2 = data[2]
```

```python
x
```

```python
plt.bar(x, y0, color = 'b', width = 0.5)
plt.bar(x+.5,y1, color = 'g', width = 0.5)
plt.bar(x+1,y2, color = 'r', width=0.5)
```

```python
plt.bar(x, y0, color = 'b', width = 0.5)
plt.bar(x, y1, color='g', width=0.5, bottom=np.sum(data[:0], axis=0))
plt.bar(x, y2, color ='r',width =0.5,bottom=np.sum(data[:1], axis=0))
```

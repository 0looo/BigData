```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
# url 설정 (1~ 100페이지)
# for page in range(100) :  
# url = 'https://movie.naver.com/movie/point/af/list.nhn?&page={}'.format(page)
url = 'https://movie.naver.com/movie/point/af/list.nhn?&page=1'    # 우선 1페이지로 설정
res = requests.get(url)
html = res.text
soup = bs(html, 'html.parser')
```

```python
# 영화제목, 평점, 후기 전체 t_path
#old_content > table > tbody > tr:nth-child(1) > td.title
t_path = soup.select('td.title')
# t_path[0]   0번째 후기(전체포함)
```

```python
# 영화제목 추출
t_path[2].select('a.movie')[0].text    # 2번째 리뷰 영화제목
```

```python
# 평점 추출
t_path[2].select('div.list_netizen_score')[0].text  # 2번째 리뷰 평점추출
```

```python
# 후기 내용
t_path[4].text.split('\n\n')[2].split('\n\t')[0]    # 3번째 후기 추출, \n신고\n 나오면 미작성
```

```python
for num in range(10) :
    url = 'https://movie.naver.com/movie/point/af/list.nhn?&page={}'.format(num)
    res = requests.get(url)
    html = res.text
    soup = bs(html, 'html.parser')
    t_path = soup.select('td.title')
    print('''---------{}번째 페이지 리뷰---------'''.format(num+1))

    for i in range(10):
        title = t_path[i].select('a.movie')[0].text
        star = t_path[i].select('div.list_netizen_score')[0].text
        text = t_path[i].text.split('\n\n')[2].split('\n\t')[0]
        if text == '\n신고\n' :
            text = '미작성'
        print('''{}번 리뷰제목 : {}{}후기 : {}
        '''.format(i+1,title,star,text))
```

```python
from bs4 import BeautifulSoup
import requests  # 웹사이트의 html script를 긁어오는 라이브러리

url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0'

html = requests.get(url)
print(html.text)
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
#newsColl > div.coll_cont
```

```python
res = soup.select('#newsColl > div.coll_cont > ul > li > div > div > div > a')
res
```

```python
for i,t in enumerate(res) :
    print(i+1, ':', t.text)
```

```python
for i,t in enumerate(res) : 
    print(i+1, ':',t['href'])
```

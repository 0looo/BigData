```python
def K_P() :
    Key = input('키워드를 입력하세요 : ')
    P = input('검색할 페이지수를 입력하세요 : ')
    print('검색결과 -------------- ')
    for i in range(int(P)):
        url = 'https://search.daum.net/search?nil_suggest=btn&w=news&DA=PGD&cluster=y&q={}&p={}'.format(Key,i)
        html = requests.get(url)
        soup = BeautifulSoup(html.text,'html.parser')
        txt = soup.select('a.f_link_b')
        txt2 = soup.select('span.f_nb')
        txt3 = soup.select('p.f_eb')
        
        title_list = []
        url_list = []
        date_list = []
        factory_list = []
        summary_list = []
        
        for t in range(len(txt)):
            title_list.append((txt[t])
            url_list.append(txt[t]['href'])
            date_list.append(txt2[t].text.split('|')[0])
            factory_list.append(txt2[t].text.split('|')[1])
            summary_list.append(txt3[t].text)
            
        for i in range(len(txt)) :
            print('''============================================================
제목 : {}
URL : {}
날짜 : {}
언론사 : {}
summary : {}
============================================================='''.format(txt[i].text,txt[i]['href'],txt2[i].text.split('|')[0],txt2[i].text.split('|')[1],txt3[i].text))
        return title_list, url_list, date_list, factory_list, summary_list
```

```python
# 다음 키워드로 검색한 뉴스페이지에서 1페이지부터 page_num페이지까지 모든 뉴스를 크롤링

def daum_news_crawling(keyword, page_num):
    
    title_list = []
    url_list = []
    date_list = []
    summary_list = []
    
    count = 0
    for page in range(1, page_num+1):
        url = 'https://search.daum.net/search?w=news&DA=PGD&enc=utf8&cluster=y&cluster_page=10&q={}&p={}'
        url = url.format(keyword, page)
        print(url)

        html = requests.get(url)

        soup = BeautifulSoup(html.text, 'html.parser')
        res_t = soup.select('a.f_link_b')

        path = '#clusterResultUL > li > div.wrap_cont > div'
        res = soup.select(path)
        
        for t in res:
            title_list.append(t.select('a.f_link_b')[0].text)
            url_list.append(t.select('a.f_link_b')[0]['href'])
            date_list.append(t.select('span.f_nb.date')[0].text)
            summary_list.append(t.select('div.wrap_cont > div > p')[0].text) 

        for t in res:
            print('='*30)
            count += 1
            print('crawling num :', count)
            print('제목 :', t.select('a.f_link_b')[0].text)
            print('URL :',t.select('a.f_link_b')[0]['href'])
            print('Date :',t.select('span.f_nb.date')[0].text)
            print('Summary :',t.select('div.wrap_cont > div > p')[0].text) 
            print()
            
    return title_list, url_list, date_list, summary_list
```

```python
title_list, url_list, date_list, summary_list = daum_news_crawling('머신러닝', 10)
```

```python
daum_news_dict = {}

daum_news_dict['title'] = title_list
daum_news_dict['url'] = url_list
daum_news_dict['date'] = date_list
daum_news_dict['summary'] = summary_list

daum_news_dict
```

```python
import pandas as pd
df = pd.DataFrame(daum_news_dict)
df.head()         # 상위 5개 항목 표시
```

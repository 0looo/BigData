#### daum에서 검색어를 입력하고, 뉴스 섹션에서 첫페이지에 제목, 본문링크, 언론사, summery

```python
#형식
# -------------
# 제목 : 
# URL : 
# 언론사 : 
# summary :
# -------------
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EA%B0%9C%EB%AF%B8%EB%8A%94%20%EC%98%A4%EB%8A%98%EB%8F%84%20%EB%9A%A0%EB%9A%A0'
html = requests.get(url)
print(html.text)
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
#clusterResultUL
```

```python
#clusterResultUL > li.fst > div.wrap_cont > div > div.wrap_tit.mg_tit > a
```

```python
# 제목, url 추출
res = soup.select('#clusterResultUL > li > div > div > div > a')
res[0]['href']
```

```python
# 날짜 추출
res2 = soup.select('#clusterResultUL > li > div > div > span')
l1 = [res2[i].text.split('|')[0] for i in range(len(res2)) if (i % 3) == 1]
print(l1)
# 언론사 추출
res2 = soup.select('#clusterResultUL > li > div > div > span')
l2 = [res2[i].text.split('|')[1] for i in range(len(res2)) if (i % 3) == 1]
print(l2)
```

```python
# summary 추출
res3 = soup.select('#clusterResultUL > li > div > div > p')
res3[0].text
```

```python
res = soup.select('#clusterResultUL > li > div > div > div > a')
for i in range(len(res)):
    print('''
============================================================
제목 : {}
URL : {}
날짜 : {}
언론사 : {}
summary : {}
============================================================='''.format(res[i].text, res[i]['href'], l1[i], l2[i],res3[i]))
```

#### 함수로 만들어서 매개변수로 검색 키워드, 크롤링할 페이지를 입력받고  Daum 뉴스에서 키워드 검색결과 입력된 페이지 만큼 크롤링하는 코드

```python
from bs4 import BeautifulSoup
import requests
```

```python
# url 설정
def K_P() :
    Key = input('키워드를 입력하세요 : ')
    P = input('검색할 페이지수를 입력하세요 : ')
    print('검색결과 -------------- ')
    for i in range(int(P)):
        url = 'https://search.daum.net/search?nil_suggest=btn&w=news&DA=PGD&cluster=y&q={}&p={}'.format(Key,i)
        html = requests.get(url)
        soup = BeautifulSoup(html.text,'html.parser')
        txt = soup.select('a.f_link_b')
        for i in txt :
            print('''뉴스제목 : {}
URL : {}'''.format(i.text,i['href']))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### sklearn으로 dataset 을 생성

```python
# make_blobs : clustering sample 데이터를 생성하는 함수.
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2,
        centers=4, cluster_std = 1.8, random_state=111)
```

```python
data
```

```python
x = data[0][:, 0]
y = data[0][:, 1]
```

```python
data[1]
```

```python
plt.scatter(x,y, c=data[1], cmap='rainbow')
```

##### clustering

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(data[0])
```

```python
# center points

km.cluster_centers_
```

##### clustering result

```python
km.labels_
```

```python
# original data vs kmeans clustered data
```

```python
km = KMeans(n_clusters=4)
km.fit(data[0])

f, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('Original')
ax1.scatter(x,y, c=data[1], cmap='rainbow')

ax2.set_title('KMeans')
ax2.scatter(x,y, c=km.labels_, cmap='rainbow')
```

```python
km = KMeans(n_clusters=4)
km.fit(data[0])

f, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('Original')
ax1.scatter(x,y, cmap='rainbow')

ax2.set_title('KMeans')
ax2.scatter(x,y, c=km.labels_, cmap='rainbow')
```

```python
km = KMeans(n_clusters=2)
km.fit(data[0])

f, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('Original')
ax1.scatter(x,y, cmap='rainbow')

ax2.set_title('KMeans')
ax2.scatter(x,y, c=km.labels_, cmap='rainbow')
```

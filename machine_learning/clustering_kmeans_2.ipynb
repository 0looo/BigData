```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/College_Data", index_col=0)
df.head()
```

```python
private_data = df['Private'].replace({'Yes':1, "No":0})
private_data
```

##### 미국대학교 정보 데이터를 기반으로 n_clusters = 2, clustering한 후 사립과 공립으로 clustering이 되었는지 확인
##### 여기서는 실제 데이터와 kmeans가 clustering한 데이터의 정확도를 비교해서 봐도 됩니다..

```python
df = df.drop('Private', axis=1)
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
km.fit(df)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
private_data.values
```

```python
np.mean(private_data.values == km.labels_)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(private_data.values, km.labels_))
```

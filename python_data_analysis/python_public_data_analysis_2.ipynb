### 지역 이름과 나이, 성별을 입력 받은 후 해당 인구수를 출력

#### 남성인구수가 가장 많은 동, 여성 인구수가 가장 많은 동 출려

```python
f = open('./python_da_files/gender.csv', 'r')      # 여성은 man_age -> woman_age / 1 -> 105 나머지 동일
data = csv.reader(f)
man_age = []

for line in data : 
    if '동' == line[0].split('(')[0][-1] :
        man_age.append(int(line[1]))

f = open('./python_da_files/gender.csv', 'r')
data = csv.reader(f)
for line in data : 
    try :
        int(line[1])
    except ValueError :
        continue
    else:
        if int(line[1]) == max(man_age) : 
            print('남성이 가장 많은 동은 {}이며, 인구는 {}명 입니다.'.format(line[0], max(man_age)))
```

##### am10~ am12시 사이에 승차인원이 가장 많은 역은?

```python
f = open('./python_da_files/subwaytime.csv',encoding = 'cp949')
data = csv.reader(f)
next(data)
next(data)

s_d = {}

for line in data :
    sum_p = int(line[16])+int(line[18])
    s_d[sum_p] = line[3]
    
print('승차인원이 가장 많은 역 : {} \n승차인원 : {}'.format(s_d[max(s_d.keys())],max(s_d.keys())))
```

#### 시간을 입력받은 후 해당시간에 승차인원이 가장 많은 역은?

```python
import csv
f= open('./python_da_files/subwaytime.csv',encoding = 'cp949')
data = csv.reader(f)
next(data)
next(data)

num = int(input('시간을 입력하세요 : '))
d_t = {}

for line in data :
    d_t[int(line[2*num-4])] = line[3]

# print(d_t)
print(d_t[max(d_t.keys())], max(d_t.keys()))
```
